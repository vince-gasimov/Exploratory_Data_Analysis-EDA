### b. Movies and actors who have worked in more than one movie 
Visualize the top 20 actor according to their numbers of movies.

````python
movies_df.shape
````

The **Counter** class is imported from the collections module. Counter is used to count the number of times a hashable object appears, i.e. it is a frequency counter.

**actors_counter = Counter():** Here, an empty Counter object named actors_counter is created.

**for actors in movies['actors']:
    actors_list = [actor.strip() for actor in actors.split(',')]
    actors_counter.update(actors_list)**

This loop is applied to each row (movie) in the “actors” column of the movies DataFrame. The actor names in each row are separated by commas, then each name is striped (strip()) and a list is inserted. Then, the actors_counter object is updated with this list, i.e. the number of appearances of each actor is incremented.

**top_actors = actors_counter.most_common(20)**: The most_common method sorts the elements in actors_counter from most common to least common and returns the top 20. top_actors becomes a list and each element is a tuple: (actor_name, movie_number)

Shows the top 20 actors who have appeared in the most movies and the number of movies they have starred in.

````python
movies_df.head()
````

````python
movies_df['actors']
````

````python
from collections import Counter

actors_counter = Counter()

for actors in movies_df['actors']:
    actors_list = [actor.strip() for actor in actors.split(',')]
    actors_counter.update(actors_list)


top_actors = actors_counter.most_common(20)
top_actors
````

````python
top_actors_df = pd.DataFrame(top_actors, columns=['Actors', '#Movie_Starred'])
top_actors_df
````

````python

plt.figure(figsize=(12,10))

ax = sns.barplot(x='#Movie_Starred', y='Actors', data=top_actors_df, palette='RdPu_r')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=1)
sns.despine(left=True)

plt.title('Top 20 actors based on number of movies', fontsize=16)
plt.xlabel('Actors', fontsize=14)
plt.ylabel('Number of films', fontsize=14)

plt.xticks(fontsize=12, rotation=0)

plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
````
